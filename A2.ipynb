```python
import numpy as np

# Defining the function and its derivative
f = lambda x: np.sin(x)
df = lambda x: np.cos(x)

# Defining the finite difference approximation
D = lambda x, dx, alpha1, alpha2, alpha3, beta: (1/dx) * (alpha1*f(x-beta*dx) + alpha2*f(x) + alpha3*f(x+beta*dx))

# Seting the step size and the point at which to approximate the derivative
dx = 0.01
x = 0.5

# Seting the values of alpha1, alpha2, alpha3
alpha1 = -1/(2*beta)
alpha2 = 0
alpha3 = -alpha1

# Calculating the error for different values of beta
betas = [0.5, 1, 1.5]
for beta in betas:
    approx_derivative = D(x, dx, alpha1, alpha2, alpha3, beta)
    exact_derivative = df(x)
    error = abs(approx_derivative - exact_derivative)
    print(f"For beta = {beta}, the error is {error}")
```
